# Plotting

```python
gender_counts = df['sex'].value_counts()
```

```python
gender_counts.plot(kind='pie',
                   autopct='%1.1f%%',
                   startangle=140)
```

**Dapat dilihat dari visualisasi di atas memberikan informasi bahwa penumpang kapal titanic mayoritas-nya adalah pria**

```python
sns.histplot(df['age'].dropna(), bins=20, kde=True)
plt.title('Distribusi Umur Penumpang Kapal Titanic')
plt.xlabel('Umur Penumpang Kapal Titanic')
plt.ylabel('Jumlah Penumpang Kapal Titanic')
```

Dapat dilihat bahwa persebaran data umur penumpang berdistribusi normal

Tiket mana aja sih yang paling laku?

```python
sns.countplot(x='class', data=df, palette='Set2')
plt.title('Jumlah penumpang per Kelas Tiket')
plt.xlabel('Kelas')
plt.ylabel('Jumlah Penumpang')
```

dapat dilihat dari visualisasi data diatas bahwa jumlah penumpang dikelas 3 adalah yang paling banyak diantara kelas yang lainnya

```python
sns.boxplot(x='class', y='fare', data=df, palette='pastel')
plt.title('Boxplot antar class dan juga fare')
```

Dapat dilihat dari visualisasi diatas bahwa terdapat nilai anaomali pada kelas satu
